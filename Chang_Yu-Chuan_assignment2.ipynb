## 7) How many players in the 1960s (yearID) have hit 200 or more HRs? (Dataframe, first name, last name, number of HRs, ordered descending by HR amt)

##  8) Who has hit the most HRs in history? (Dataframe, first name, last name, HRs)

```python
playerlist8 = df[['playerID','nameFirst','nameLast']]
q8 = df[['playerID','HR']]
q88 = q8.groupby('playerID',as_index = False).sum()
q888 = pd.merge(q88,playerlist8).drop_duplicates()
q888[q888['HR']==q888['HR'].max()][['nameFirst','nameLast','HR']]
```

## 9)	Who had the most hits (H) in the 1970s? (Dataframe)

```python
playerlist9 = df[['playerID','nameFirst','nameLast']]
q9 = df.loc[(df['yearID']>=1970) & (df['yearID'] <= 1979), ['playerID','H']]
q99 = q9.groupby('playerID', as_index = False).sum() 
q999 = pd.merge(q99,playerlist9).drop_duplicates()
q999[q999['H']==q999['H'].max()][['nameFirst','nameLast','H']]
```

## 10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe, first name, last name, OBP, ordered descending by OBP)

```python
df['OBP'] = (df['H'] + df['BB'] + df['HBP'])/(df['AB']+ df['BB'] + df['HBP'] + df['SF'])
df['PA'] = df['AB'] + df['BB'] + df['HBP'] + df['SF'] + df['SH']
df[(df['PA']>=500) & (df['yearID'] == 1977)][['nameFirst','nameLast','OBP']].sort_values('OBP', ascending = False).head(5)
```

## 11)	Top 8 highest averages in 2013 with at least 300 PAs? (Dataframe, first name, last name, average, descending by average)

```python
df['Average'] = df['H']/df['AB']
df[(df['PA']>=300) & (df['yearID'] == 2013)][['nameFirst','nameLast','Average']].sort_values('Average',ascending = False).head(8)
```

## 12)	Leaders in hits from 1940 up to and including 1949. (Dataframe, first name, last name, number of hits) (top 5)

```python
playerlist12 = df[['playerID','nameFirst','nameLast']]
hitsum = df[['playerID','H','yearID']]
q12 = hitsum[(hitsum['yearID']>=1940) & (hitsum['yearID']<=1949)].groupby('playerID',as_index = False).sum().drop('yearID',1)
pd.merge(q12,playerlist12)[['nameFirst','nameLast','H']].drop_duplicates().sort_values('H',ascending = False).head(5)
```

## 13)	Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number of hits)

```python
q13 = df.groupby(['playerID','yearID'],as_index=False)[['H']].sum()
q13.groupby(['yearID']).max().groupby('playerID').count().sort_values('H',ascending=False).head(1)
```

## 25)	Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)

```python
df['HRCUM'] = df.groupby(['playerID'])['HR'].cumsum()
df[df['HRCUM']>=300].sort_values('age')[['nameFirst','nameLast','yearID']].head(5)
```
